```python
import pandas as pd
import numpy as np
import os
import csv
```

```python
# add the path of data
path = './part2'
output_path = './drop_name_data'
```

```python
def list_folders_files(path):
    """
    返回 "文件夹" 和 "文件" 名字
    :param path: "文件夹"和"文件"所在的路径
    :return:  (list_folders, list_files)
            :list_folders: 文件夹
            :list_files: 文件
    """
    list_folders = []
    list_files = []
    for file in os.listdir(path):
        file_path = os.path.join(path, file)
        if os.path.isdir(file_path):
            list_folders.append(file)
        else:
            list_files.append(file)
    return (list_folders, list_files)
```

```python
def getAlltxtFiles(file_names):
    new_file_list = []
    for i in file_names:
        if('txt' in i):
            new_file_list.append(i)
    return new_file_list
```

```python

folders_, names_ = list_folders_files(path)
file_names = getAlltxtFiles(names_) # all txt files
```

```python
def changePrefix(file_name):
    parts = file_name.split('.')
    new_name = parts[0] + '.csv'
    return new_name
```

```python
def txt2csv(inputfile): 
    agg_heads = ['inside_baoan', 'inside_10km_baoan', 'inside_city']
    with open(inputfile, encoding='ISO-8859-1') as f:
        head = f.readline()
        heads = head.split(',')[:7]
        datas = f.readlines()
        
        new_data = []
        for data in datas:
            data_ = data.split(',')[:7]
            jd = data_[2]
            wd = data_[3]
            poi = (float(jd), float(wd))
#             print(poi)
            result = writeCarStatusAirport(poi)
            data_ = data_ + result
            new_data.append(data_)
        
        heads += agg_heads
        df = pd.DataFrame(new_data, columns=heads)
        df = df.drop(axis=1, columns=['name'])
        return df
```

```python
cnt = 0
for file_name in file_names:
    cnt += 1
    file_read_path = path + '/' + file_name

    file_output_path = output_path + '/' + changePrefix(file_name)
    # file_output_path, file_read_path
    df = txt2csv(file_read_path)
    # df
    df.to_csv(file_output_path,index=False)
    print('output', file_output_path, cnt)
# txtDF = pd.DataFrame(txt)

# txtDF.to_csv(file_output_path, index=False)
```
