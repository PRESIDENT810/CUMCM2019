```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
```

```python
%matplotlib inline
```

```python
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
def start_location(series):
    jd = series['StartJD']
    wd = series['StartWD']
    point = Point(jd, wd)
#     bao an
    vertex_lst_baoan = [
        [113.793451,22.656175],
        [113.801069,22.659776],
        [113.803225,22.653506],
        [113.815585,22.660177],
        [113.818783,22.655374],
        [113.826077,22.659043],
        [113.847062,22.627022],
        [113.817454,22.614612]
    ]
    
    baoan = Polygon(vertex_lst_baoan)
    result_baoan = baoan.contains(point)
    
    return result_baoan

def end_location(series):
    jd = series['EndJD']
    wd = series['EndWD']
    point = Point(jd, wd)
#     bao an
    vertex_lst_baoan = [
        [113.793451,22.656175],
        [113.801069,22.659776],
        [113.803225,22.653506],
        [113.815585,22.660177],
        [113.818783,22.655374],
        [113.826077,22.659043],
        [113.847062,22.627022],
        [113.817454,22.614612]
    ]
    
    baoan = Polygon(vertex_lst_baoan)
    result_baoan = baoan.contains(point)
    
    # bao an 10km inside
    vertex_l0km_baoan = [
        [113.803734,22.563633],
        [113.893995,22.608414],
        [113.840815,22.711135],
        [113.737474,22.692198],
    ]
    
    baoan_10km = Polygon(vertex_l0km_baoan)
    result_baoan_10km = baoan_10km.contains(point)
    
    result_city = False
    vertex_city = [
        [113.909374,22.620691],
        [113.892127,22.522442],
        [114.147389,22.52992],
        [114.13934,22.600407]
    ]
    
    
    city = Polygon(vertex_city)
    result_city = city.contains(point)
    
    
    return [result_baoan, result_baoan_10km, result_city]
```

```python
def add_locationInfo(row):
    global location_df
    FromAirport = start_location(row)
    end_info = end_location(row)
    InAirport = end_info[0]
    InNearby = end_info[1]
    Faraway = end_info[2]
    location_df.loc[location_df.shape[0]+1] = {'FromAirport': FromAirport, 'InNearby': InNearby, 'Faraway': Faraway}
```

```python
filenames = os.listdir('./TripData/')
```

```python
trip_df = pd.read_csv('./TripData/%s'%filenames[0], index_col=0)
alltrip_df = trip_df[0:0]

for filename in filenames:
    location_df = pd.DataFrame({'FromAirport':[], 'InNearby': [], 'Faraway': []})
    trip_df = pd.read_csv('./TripData/%s'%filename, index_col=0)
    trip_df.apply(add_locationInfo,axis=1)
    trip_df = pd.concat([trip_df,location_df],axis=1)
    trip_df = trip_df[trip_df['FromAirport']==1]
    alltrip_df = alltrip_df.append(trip_df)
```
