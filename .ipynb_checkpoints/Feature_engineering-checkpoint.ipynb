```python
import pandas as pd
import numpy as np
import matplotlib.mlab as mlab 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./useable_data.csv',index_col=0)
```

```python
df_copy = df.copy()
```

```python
def get_daytime(row):
    hour = int(row['hour'])
    if hour<6.5:
        return 'EarlyMorning'
    elif hour<12.5:
        return 'Morning'
    elif hour<18.5:
        return 'Afternoon'
    else:
        return 'Evening'
```

```python
df.head()
```

```python
df['DayTime'] = df.apply(get_daytime,axis=1)
df = df.drop(labels=['CarID','StartTime','hour','AvgSpeed'],axis=1)
```

```python
df = pd.get_dummies(df)
```

### Most waiting time is quite short

```python
plt.figure(figsize=(12,8))

pick_df = df[df['Pick']==1]
pick = pick_df['WaitingTime'].value_counts()

leave_df = df[df['Pick']==0]
leave = leave_df['WaitingTime'].value_counts()

plt.bar(leave.index,pick[leave.index],color="red",label="Pick up")
plt.bar(leave.index,leave.values,bottom=pick[leave.index],color="blue",label="Not pick up")
plt.title("WaitingTime Info")
# plt.xlim=500
plt.xlabel("WaitingTime")
plt.ylabel("Count")
plt.legend(loc=1)
```

### Drivers who pick up passengers are likely to wait for a short time

```python
plt.figure(figsize=(12,12))

Evening_pick = df[ (df["Pick"]==1) & (df["DayTime_Evening"]==1)].shape[0]
Evening_leave = df[ (df["Pick"]==0) & (df["DayTime_Evening"]==0)].shape[0]
Afternoon_pick = df[ (df["Pick"]==1) & (df["DayTime_Afternoon"]==1)].shape[0]
Afternoon_leave = df[ (df["Pick"]==0) & (df["DayTime_Afternoon"]==0)].shape[0]
Morning_pick = df[ (df["Pick"]==1) & (df["DayTime_Morning"]==1)].shape[0]
Morning_leave = df[ (df["Pick"]==0) & (df["DayTime_Morning"]==0)].shape[0]
Early_pick = df[ (df["Pick"]==1) & (df["DayTime_EarlyMorning"]==1)].shape[0]
Early_leave = df[ (df["Pick"]==0) & (df["DayTime_EarlyMorning"]==0)].shape[0]

plt.subplot(2,2,1)
plt.pie([Evening_pick,Evening_leave],colors=('green','indigo'),labels=("Pick","Leave"), autopct='%1.2f%%')
plt.title("Evening Pick Stats")
plt.legend(loc=0)

plt.subplot(2,2,2)
plt.pie([Afternoon_pick,Afternoon_leave],colors=('green','indigo'),labels=("Pick","Leave"), autopct='%1.2f%%')
plt.title("Afternoon Pick Stats")
plt.legend(loc=0)

plt.subplot(2,2,3)
plt.pie([Morning_pick,Morning_leave],colors=('green','indigo'),labels=("Pick","Leave"), autopct='%1.2f%%')
plt.title("Morning Pick Stats")
plt.legend(loc=0)

plt.subplot(2,2,4)
plt.pie([Early_pick,Early_leave],colors=('green','indigo'),labels=("Pick","Leave"), autopct='%1.2f%%')
plt.title("Early Morning Pick Stats")
plt.legend(loc=0)
```

### Analyse the distribution of waiting time

```python
### waiting time belongs to exponential distribition
wait_mean = np.mean(df["WaitingTime"])
wait_mean
```

```python
fig = plt.figure(figsize=(12,8))
num_bins = 30

lam = 1/wait_mean
x = np.arange(0,100,0.1)
y = lam * np.exp(- lam * x)
n, bins, patches = plt.hist(df['WaitingTime'], num_bins,normed=1, facecolor='blue', alpha=0.5)
plt.plot(x,y,color='orange', lw=3)
```
